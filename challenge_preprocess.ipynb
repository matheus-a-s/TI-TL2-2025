```python
import pandas as pd

data_raw = pd.read_excel("./dataset/challenge/original/DataSet_18_10_2023.xlsx")
data_raw
```

```python
data = data_raw.dropna()

data
data.to_csv('./dataset/challenge/DataSet_Without_NaN.csv')
```

```python
data_processed = data[~((data['attendance_T'] == 0) | 
                   (data['attendance_TP'] == 0) | 
                   (data['attendance_PL'] == 0) |
                    (data['frequency_grade'] == 0))]
```

```python
data_processed.to_csv('./dataset/challenge/DataSet_Attendance.csv')
data_processed
```

```python
df = pd.read_csv('./dataset/challenge/DataSet_Attendance.csv', header=0, index_col=0)
df
```

```python
colunas = ['dropout']

for col in colunas:
    print(f"\nAnalisando a coluna: {col}")
    
    # Contagem de valores Ãºnicos
    contagem = data_processed[col].value_counts(dropna=False)
    print("Contagem:\n", contagem)
    
    # Porcentagem
    porcentagem = data_processed[col].value_counts(normalize=True, dropna=False) * 100
    print("Porcentagem (%):\n", porcentagem)
```

```python
df_dropout_true = data_processed[data_processed['dropout'] == True]

df_dropout_false = data_processed[data_processed['dropout'] == False]

df_dropout_true.to_excel('./dataset/challenge/dropout_true.xlsx', index=False)
df_dropout_false.to_excel('./dataset/challenge/dropout_false.xlsx', index=False)

print("Arquivos salvos: dropout_true.csv e dropout_false.csv")
```
